```python
from pymongo import MongoClient
client = MongoClient("localhost", 27017)
rss_db = client.seattle_news.sea_news_rss
```

```python
council_links = []
for rss in rss_db.find():
    link = rss["link"]
    if "council.seattle.gov" in link:
        council_links.append(link)

council_links = list(set(council_links))
council_links
```

```python
raw_dat = []
for _, link in enumerate(council_links):
    page = rq.get(link, headers = headers)
    soup = bs(page.content)
    content = soup.find_all("div",  {"class": "content entry"})
    if len(content) > 0:
        print(f"scraping link {_}: {link}")
        text = content[0].text
        newobj = {"link": link, "text": text}
        raw_dat.append(newobj)
```

```python
council_arts_db.insert_many(raw_dat)
```

```python
db = client.seattle_news
for doc in council_arts_db.find():
    text = doc["text"]
    cfinds = re.findall("Councilmember [A-Z]\W[A-Z][a-z]*|Councilmember [A-Z][a-z]*", text)
    rfinds = re.findall("[A-Z]*[a-z]*, News Releases", text)
    if len(rfinds) == 1:
        name = rfinds[0].split(", ")[0]
        council_arts_db.update_one({"_id": doc["_id"]}, {"$set": {"involved_councilmembers": name}})
    elif len(cfinds) > 0:
        name = cfinds[0].split(" ")[1]
        council_arts_db.update_one({"_id": doc["_id"]}, {"$set": {"involved_councilmembers": name}})
    else:
        council_arts_db.update_one({"_id": doc["_id"]}, {"$set": {"involved_councilmembers": "unknown"}})
```

```python
from datetime import datetime as dt
for doc in council_arts_db.find():
    link = doc["link"]
    date = re.findall("[0-9]{4}/[0-9]{2}/[0-9]{2}", link)
    if len(date) > 0:
        date = dt.strptime(date[0], "%Y/%m/%d")
        council_arts_db.update_one({"_id": doc["_id"]}, {"$set": {"pub_date": date}})
```

```python
table = {
    "link": [],
    "date": [],
    "text": [],
    "councilmember": []
}
for doc in council_arts_db.find():
    table["link"].append(doc["link"])
    table["date"].append(doc["pub_date"])
    table["text"].append(doc["text"])
    table["councilmember"].append(doc["involved_councilmembers"])
```

```python
import polars as pl
df = pl.DataFrame(table)
```

```python
from polars import col as c
df = df.with_columns(c("text").str.to_lowercase())
df
```

```python
escapes = ''.join([chr(char) for char in range(1, 32)])
df = df.with_columns(c("text").map_elements(lambda x: x.translate(str.maketrans("", "", escapes))))
```

```python
import nltk
from nltk.corpus import stopwords
import re
en_stop = stopwords.words("english")
no_sw = []
for row in df.iter_rows(named = True):
    text = row["text"]
    text = "".join(re.split("[^ A-Za-z]", text))
    no_sw.append(text)
    tokens = nltk.word_tokenize(text)
    tokens = [w for w in tokens if w not in en_stop]
    fd = nltk.FreqDist(tokens)
    print(fd.most_common(5))
```

```python
df = df.with_columns(pl.Series(
    name = "no_sw",
    values = no_sw
))
df.write_csv("council_ann.csv")
```

```python
with open("possible_categories.txt", "r") as file:
    cats = file.read().split("\n")
```

```python
import polars as pl
category_string = ", ".join(cats)
df = pl.read_csv("council_ann.csv")
df
```

```python
from FlagEmbedding import FlagReranker
rr = FlagReranker('BAAI/bge-reranker-v2-m3')
```

```python
counter = 1
scores = {cat: [] for cat in cats}
for row in df.iter_rows(named = True):
    print(counter)
    counter += 1
    for cat in cats:
        scores[cat].append(rr.compute_score([cat, row["text"]], normalize = True))
scores = pl.DataFrame(scores)
scores.to_pandas()
```

```python
scores_df = pl.concat([df.select(["date", "councilmember"]), scores], how = "horizontal")
scores_df.write_csv("seattle_council_announcement_scores.csv")
```
