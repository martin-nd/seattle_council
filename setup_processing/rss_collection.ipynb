```python
entries = ["notempty"]
feed_url = "https://news.seattle.gov/feed/?paged="
counter = 1
pages = []
while len(entries) != 0:
    print(f"scraping page {counter}")
    page_url = feed_url + str(counter)
    page = fp.parse(page_url)
    entries = page.entries
    pages.append(page)
    counter += 1
```

```python
from pymongo import MongoClient
client = MongoClient("localhost", 27017)
```

```python
col = client.seattle_news.sea_news_rss
```

```python
for page in pages:
    if len(page.entries) != 0:
        col.insert_many(page.entries)
```
